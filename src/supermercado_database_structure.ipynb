# **SUPERMERCADO DATABASE structure.**

## **Importar librerías.**

```python
import pandas as pd
```

## **Cargar datos.**

```python
categorias = pd.read_csv('../data/categorias_fyv.csv')
subcategorias = pd.read_csv('../data/subcategorias_fyv.csv')
productos = pd.read_csv('../data/productos_fyv.csv')
```

## **Generar estructura.**

### **Secciones.**

```python
# crear tabla "Secciones" con "Fruta y Verdura" como sección
secciones = pd.DataFrame({'seccion_id' : ['1'], 'seccion' : ['Fruta y Verdura']})

secciones
```

### **Categorías.**

```python
# insertar columna de IDs llamada "seccion_id", para realizar la correspondiente relación en la base de datos como "FOREIGN KEYS"
categorias.insert(len(categorias.columns), 'seccion_id', 1)

categorias
```

### **Subcategorías.**

```python
# insertar columna de IDs llamada "seccion_id", para realizar la correspondiente relación en la base de datos como "FOREIGN KEYS"
subcategorias.insert(len(subcategorias.columns), 'seccion_id', 1)

subcategorias
```

### **Productos.**

```python
# insertar columna de IDs llamada "seccion_id", para realizar la correspondiente relación en la base de datos como "FOREIGN KEYS"
productos.insert(len(productos.columns), 'seccion_id', 1)

productos
```

### **Ventas.**

```python
# crear tabla "ventas", con columnas "venta_id", "fecha", "seccion_id"
ventas = pd.DataFrame(columns=['venta_id', 'fecha', 'dia_semana', 'seccion_id'])

ventas
```

### **Ventas-Productos.**

```python
# crear tabla "ventas_productos", con columnas "venta_id", "producto_id", "cantidad"
ventas_productos = pd.DataFrame(columns=['venta_id', 'producto_id', 'cantidad'])

ventas_productos
```

## **Exportar tablas.**

```python
# exportar tablas transformadas a ".csv"
secciones.to_csv('../data/secciones_fyv.csv', index=False)

categorias.to_csv('../data/categorias_fyv.csv', index=False)

subcategorias.to_csv('../data/subcategorias_fyv.csv', index=False)

productos.to_csv('../data/productos_fyv.csv', index=False)

ventas.to_csv('../data/ventas_fyv.csv', index=False)

ventas_productos.to_csv('../data/ventas_productos_fyv.csv', index=False)
```
