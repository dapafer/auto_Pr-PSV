# **MERCADONA SCRAPING FyV**

## **Importar librerías**

```python
import warnings
warnings.filterwarnings('ignore')

import time

import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# sustituye al archivo
from webdriver_manager.chrome import ChromeDriverManager

# para buscar por tag, clase, id...
from selenium.webdriver.common.by import By

# para pulsar teclas
from selenium.webdriver.common.keys import Keys

# para esperar a que cargue
#from selenium.webdriver.support.ui import WebDriverWait

# condiciones esperadas
#from selenium.webdriver.support import expected_conditions as EC
```

## **Scraping**

```python
# instalar driver de chrome
PATH = ChromeDriverManager().install()

# driver configuration
opciones = Options()

# quitar bandera de ser robot
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)

# si True, no aparece la ventana (headless=no visible)
opciones.headless = True

# comenzar maximizado
opciones.add_argument('--start-maximized')

# mantener cookies
#opciones.add_argument('user-data-dir=selenium')

# adblocker
#opciones.add_extension('driver_folder/adblock.crx')

# modo incógnito
opciones.add_argument('--incognito')

# guardar cookies
#opciones.add_argument('user-data-dir=cookies')
```

### **Acceso a la tienda**

```python
url = 'https://www.mercadona.es'
driver = webdriver.Chrome(PATH, options=opciones)
driver.get(url)

# aceptar cookies
driver.find_element(By.XPATH, '//*[@id="root"]/div/div/div/button[2]').click()
time.sleep(3)

# escribir texto en "Código postal"
cp = driver.find_element(By.XPATH, '//*[@id="root"]/header/div/div/form/div/input')
cp.send_keys('28045')

# entrar a la tienda
driver.find_element(By.XPATH, '//*[@id="root"]/header/div/div/form/input').click()
time.sleep(3)

# entrar a "Categorías"
driver.find_element(By.XPATH, '//*[@id="root"]/header/div[1]/nav/a[1]').click()
time.sleep(3)

# entrar a "Fruta y Verdura"
driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[1]/ul/li[17]/div/button/span/label').click()
time.sleep(3)

# extraer categoría de "Fruta y verdura"
cat_fyv = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[1]/ul/li[17]/div/button/span/label').text

cat_fyv
```

### **Fruta scraping**

```python
# extraer frutas
list_frutas = driver.find_elements(By.CSS_SELECTOR, 'h4.subhead1-r.product-cell__description-name[data-test="product-cell-name"]')
frutas = [fruta.text for fruta in list_frutas]

frutas
```

```python
# extraer formato completo de frutas
list_all_form_frutas = driver.find_elements(By.CSS_SELECTOR, 'div.product-format.product-format__size--cell')
all_form_frutas = [all_form.text for all_form in list_all_form_frutas]

all_form_frutas
```

```python
# extraer formato dividido de frutas
list_div_form_frutas = driver.find_elements(By.CSS_SELECTOR, 'span.footnote1-r')
div_form_frutas = [div_form.text for div_form in list_div_form_frutas]

div_form_frutas
```

```python
# extraer precio de frutas
list_prec_frutas = driver.find_elements(By.CSS_SELECTOR, 'div.product-price > p.product-price__unit-price')
prec_frutas = [prec_fruta.text for prec_fruta in list_prec_frutas]

prec_frutas
```

```python
# extraer subcategoría de fruta
sub_frutas = driver.find_element(By.XPATH, '//h1[@id="content" and contains(@class, "category-detail__title")]').text

sub_frutas
```

### **Cambiar subcategoría**

```python
# cambiar a subcategoría "Lechuga y ensalada preparada"
driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[1]/ul/li[17]/div/ul/li[2]/button').click()
time.sleep(3)
```

### **Ensaladas scraping**

```python
# extraer ensaladas
list_ensaladas = driver.find_elements(By.CSS_SELECTOR, 'h4.subhead1-r.product-cell__description-name[data-test="product-cell-name"]')
ensaladas = [ensalada.text for ensalada in list_ensaladas]

ensaladas
```

```python
# extraer formato completo de ensaladas
list_all_form_ensaladas = driver.find_elements(By.CSS_SELECTOR, 'div.product-format.product-format__size--cell')
all_form_ensaladas = [all_form.text for all_form in list_all_form_ensaladas]

all_form_ensaladas
```

```python
# extraer formato dividido de ensaladas
list_div_form_ensaladas = driver.find_elements(By.CSS_SELECTOR, 'span.footnote1-r')
div_form_ensaladas = [div_form.text for div_form in list_div_form_ensaladas]

div_form_ensaladas
```

```python
# extraer precio de ensaladas
list_prec_ensaladas = driver.find_elements(By.CSS_SELECTOR, 'div.product-price > p.product-price__unit-price')
prec_ensaladas = [prec_ensalada.text for prec_ensalada in list_prec_ensaladas]

prec_ensaladas
```

```python
# extraer subcategoría de ensalada
sub_ensaladas = driver.find_element(By.XPATH, '//h1[@id="content" and contains(@class, "category-detail__title")]').text

sub_ensaladas
```

### **Cambiar subcategoría**

```python
# cambiar a subcategoría "Verdura"
driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[1]/ul/li[17]/div/ul/li[3]/button').click()
time.sleep(3)
```

### **Verduras scraping**

```python
# extraer verduras
list_verduras = driver.find_elements(By.CSS_SELECTOR, 'h4.subhead1-r.product-cell__description-name[data-test="product-cell-name"]')
verduras = [verdura.text for verdura in list_verduras]

verduras
```

```python
# extraer formato completo de verduras
list_all_form_verduras = driver.find_elements(By.CSS_SELECTOR, 'div.product-format.product-format__size--cell')
all_form_verduras = [all_form.text for all_form in list_all_form_verduras]

all_form_verduras
```

```python
# extraer formato dividido de verduras
list_div_form_verduras = driver.find_elements(By.CSS_SELECTOR, 'span.footnote1-r')
div_form_verduras = [div_form.text for div_form in list_div_form_verduras]

div_form_verduras
```

```python
# extraer precio de verduras
list_prec_verduras = driver.find_elements(By.CSS_SELECTOR, 'div.product-price > p.product-price__unit-price')
prec_verduras = [prec_verdura.text for prec_verdura in list_prec_verduras]

prec_verduras
```

```python
# extraer subcategoría de verdura
sub_verduras = driver.find_element(By.XPATH, '//h1[@id="content" and contains(@class, "category-detail__title")]').text

sub_verduras
```

## **Crear y exportar tablas**

### **Productos FyV**

```python
# crear tabla con lista "frutas", "all_form_frutas" y "prec_frutas"
fruta_df = pd.DataFrame({'producto' : frutas, 'formato' : all_form_frutas, 'precio' : prec_frutas})

fruta_df
```

```python
# crear tabla con lista "ensaladas", "all_form_ensaladas" y "prec_ensaladas"
ensalada_df = pd.DataFrame({'producto' : ensaladas, 'formato' : all_form_ensaladas, 'precio' : prec_ensaladas})

ensalada_df
```

```python
# crear tabla con lista "verduras", "all_form_verduras" y "prec_verduras"
verdura_df = pd.DataFrame({'producto' : verduras, 'formato' : all_form_verduras, 'precio' : prec_verduras})

verdura_df
```

```python
# exportar tablas resultantes a ".csv"
fruta_df.to_csv('../../data/scraping/frutas.csv', index=False)

ensalada_df.to_csv('../../data/scraping/ensaladas.csv', index=False)

verdura_df.to_csv('../../data/scraping/verduras.csv', index=False)
```

```python
# crear tabla con variable "sub_frutas"
sub_frut_df = pd.DataFrame({'subcategoria': [sub_frutas]})

sub_frut_df
```

```python
# crear tabla con variable "sub_ensaladas"
sub_ens_df = pd.DataFrame({'subcategoria': [sub_ensaladas]})

sub_ens_df
```

```python
# crear tabla con variable "sub_verduras"
sub_verd_df = pd.DataFrame({'subcategoria': [sub_verduras]})

sub_verd_df
```

```python
# exportar tablas resultantes a ".csv"
sub_frut_df.to_csv('../../data/scraping/sub_frutas.csv', index=False)

sub_ens_df.to_csv('../../data/scraping/sub_ensaladas.csv', index=False)

sub_verd_df.to_csv('../../data/scraping/sub_verduras.csv', index=False)
```
