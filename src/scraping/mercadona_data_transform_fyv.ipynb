## **Importar librerías**

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

## **Cargar datos**

### **Productos**

```python
# dataframes de "frutas", "ensaladas" y "verduras"
fruta = pd.read_csv('../../data/scraping/frutas.csv')

ensalada = pd.read_csv('../../data/scraping/ensaladas.csv')

verdura = pd.read_csv('../../data/scraping/verduras.csv')
```

### **Subcategorías**

```python
# dataframes de subcategorias de "Fruta y verdura"
sub_fruta = pd.read_csv('../../data/scraping/sub_frutas.csv')

sub_ensalada = pd.read_csv('../../data/scraping/sub_ensaladas.csv')

sub_verdura = pd.read_csv('../../data/scraping/sub_verduras.csv')
```

### **Categoría**

```python
# dataframe de categoría "Fruta y verdura"
cat_fyv = pd.read_csv('../../data/scraping/cat_fruta_verdura.csv')
```

## **Transformar**

### **Categoría FyV**

```python
cat_fyv
```

```python
# crear e insertar columna de IDs llamada "categoria_id", para realizar la correspondiente relación en la base de datos como "PRIMARY KEYS"
cat_fyv.insert(0, 'categoria_id', range(1, len(cat_fyv)+1))

cat_fyv
```

### **Subcategorías FyV**

```python
sub_fruta
```

```python
sub_ensalada
```

```python
sub_verdura
```

```python
# crear tabla conjunta con todas las subcategorías
sub_fyv = pd.concat([sub_fruta, sub_ensalada, sub_verdura])

#resetear índice
sub_fyv = sub_fyv.reset_index(drop=True)

sub_fyv
```

```python
# crear e insertar la columna de IDs "categoria_id", para realizar la correspondiente relación en la base de datos como "FOREIGN KEYS"
sub_fyv['categoria_id'] = 1

sub_fyv
```

```python
# crear e insertar columna de IDs llamada "subcategoria_id", para realizar la correspondiente relación en la base de datos como "PRIMARY KEYS"
sub_fyv.insert(0, 'subcategoria_id', range(1, len(sub_fyv)+1))

sub_fyv
```

```python
# crear tabla conjunta de productos
prod_fyv = pd.concat([fruta, ensalada, verdura])

#resetear índice
prod_fyv = prod_fyv.reset_index(drop=True)

prod_fyv
```

```python
# crear e insertar la columna de IDs "categoria_id", para realizar la correspondiente relación en la base de datos como "FOREIGN KEYS"
prod_fyv['categoria_id'] = 1

prod_fyv
```

## **Exportar tablas**

```python
# exportar tablas limpias y transformadas a ".csv"
cat_fyv.to_csv('../../data/fyv/categorias_fyv.csv', index=False)

sub_fyv.to_csv('../../data/fyv/subcategorias_fyv.csv', index=False)

prod_fyv.to_csv('../../data/fyv/productos_fyv.csv', index=False)
```
