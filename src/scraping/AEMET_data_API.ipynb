```python
# convertir datos a dataframe
data_dict = json.loads(data_res.content)
clima_dia_ene_may_23 = pd.DataFrame(data_dict)

clima_dia_ene_may_23
```

```python
# eliminar columnas que no aportan valor
clima_dia_ene_may_23 = clima_dia_ene_may_23.drop(columns=['indicativo', 'nombre', 'provincia', 'altitud', 'tmed', 'horatmin', 'horatmax', 'dir', 'velmedia', 'racha', 'horaracha', 'sol', 'presMax', 'horaPresMax', 'presMin', 'horaPresMin'])

clima_dia_ene_may_23
```

```python
# cambiar nombre columnas
clima_dia_ene_may_23 = clima_dia_ene_may_23.rename(columns={'prec': 'lluvia', 'tmin': 'temperatura_minima', 'tmax': 'temperatura_maxima'})

clima_dia_ene_may_23
```

```python
# modificar formato columna "fecha"
clima_dia_ene_may_23['fecha'] = pd.to_datetime(clima_dia_ene_may_23['fecha'])

clima_dia_ene_may_23
```

```python
# obtener días festivos en España para el año 2023
holidays_es = holidays.Spain(years=2023)

# definir función para verificar si una fecha es domingo o festivo
def domingo_festivo(fecha):
    return fecha.weekday() == 6 or fecha in holidays_es

# filtrar las filas que contengan domingos o festivos
clima_dia_ene_may_23 = clima_dia_ene_may_23[~clima_dia_ene_may_23['fecha'].apply(domingo_festivo)]

clima_dia_ene_may_23
```

```python
# comprobar valores columna "lluvia"
clima_dia_ene_may_23['lluvia'].unique()
```

```python
# comprobar valores columna "temperatura_minima"
clima_dia_ene_may_23['temperatura_minima'].unique()
```

```python
# comprobar valores columna "temperatura_maxima"
clima_dia_ene_may_23['temperatura_maxima'].unique()
```

```python
# modificar valor que no es numérico, convertir columnas a "float"
clima_dia_ene_may_23['lluvia'] = clima_dia_ene_may_23['lluvia'].replace('Ip', '0,1')

columns_convert = ['lluvia', 'temperatura_minima', 'temperatura_maxima']
clima_dia_ene_may_23[columns_convert] = clima_dia_ene_may_23[columns_convert].apply(lambda x: x.str.replace(',', '.'))
clima_dia_ene_may_23[columns_convert] = clima_dia_ene_may_23[columns_convert].apply(lambda x: x.astype(float))

# crear funcion para modificar los valores de la columna "precipitacion"
def modificar_precipitacion(valor):
    if valor > 0:
        return 'si'
    else:
        return 'no'

# aplicar funcion
clima_dia_ene_may_23['lluvia'] = clima_dia_ene_may_23['lluvia'].apply(modificar_precipitacion)

clima_dia_ene_may_23
```

```python
# generar rango de fechas del 1 al 30 de junio de 2023, donde se generan predicciones de venta
fechas_jun23 = pd.date_range(start='2023-06-01', end='2023-06-30', freq='D')

# crear dataframe con nuevas filas
clima_jun23 = {'fecha': fechas_jun23,
            'lluvia': 'desconocido',
            'temperatura_minima': 0,
            'temperatura_maxima': 0}
clima_dia_jun_23 = pd.DataFrame(clima_jun23)

# filtrar las filas que contengan domingos o festivos
clima_dia_jun_23 = clima_dia_jun_23[~clima_dia_jun_23['fecha'].apply(domingo_festivo)]

clima_dia_jun_23
```

```python
# concatenar dataframe "clima_dia_ene_may_23" con "clima_dia_jun_23"
clima_dia_ene_jun_23 = pd.concat([clima_dia_ene_may_23, clima_dia_jun_23], ignore_index=True)

clima_dia_ene_jun_23
```
